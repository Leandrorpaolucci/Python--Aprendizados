#### Consulta de informações de um CEP

```python
import requests

#cep sendo considerado string
cep = input("Digite o CEP:")
#tratamento de traços, pontos, espaçamentos do usuário
cep = cep.replace("-", "").replace(".", "").replace(" ", "")
numero = input("Digite o número:").isalnum()

#comparando se é igual a 8, roda o comando abaixo no if, caso contrario entra no bloco else
if len(cep) == 8:
    # link da API e modificando o cep exemplo da API por uma variavel criada acima e tratada
    link = f"https://viacep.com.br/ws/{cep}/json/"
    #transformando o link em dicionário através do requests....
    requisicao = requests.get(link)
    #pegando cada campo de informação e transformando em uma variável
    dic_requisicao = requisicao.json()
    logradouro = dic_requisicao['logradouro']
    complemento = dic_requisicao['complemento']
    bairro = dic_requisicao['bairro']
    localidade = dic_requisicao['localidade']
    uf = dic_requisicao['uf']
    ibge = dic_requisicao['ibge']
    gia = dic_requisicao['gia']
    siafi = dic_requisicao['siafi']
    
    print(f'''\033[34mO Cep inserido foi {cep} e as informações do endereço estão abaixo:\n
    Rua:{logradouro}
    Número:{numero}
    Bairro:{bairro}
    Cep:{cep}
    Cidade:{localidade}
    Estado:{uf}''')
else:
    print("\033[31mCep Errado!\nO CEP informado está errado ou não existe!")
```

#### Busca de CEP a partir de endereço

```python
import requests

def ajustar_variaveis(uf, cidade, endereco):
    #Ajustando campos Estado
    uf = input("Digite o estado:")
    uf = uf.replace(" ", "").replace(".", "").upper()
    print(f"O Estado digitado foi:{uf}")
    
    #Ajustado campos da cidade
    cidade = input("Digite a cidade:")
    cidade = cidade.replace(".", "").replace("-", "").title()
    print(f"A cidade inserida foi:{cidade}")
    
    #Ajustando campos da rua digitada
    endereco = input("Digite o nome da rua:")
    endereco = endereco.replace("-", "").replace(".", "").title()
    print(f"A rua inserida foi: {endereco}")
    

#Chamando a função com 3 variáveis
ajustar_variaveis(uf, cidade, endereco)

#Chamando a API através do link com variáveis
link_endereco = f"https://viacep.com.br/ws/{uf}/{cidade}/{endereco}/json/"

# fazendo requesição em get pegando as informações
requisicao = requests.get(link_endereco)

# transformando as informações no GET em dicionário
dic_requisicao = requisicao.json()
print(requisicao)
print(dic_requisicao)
```
